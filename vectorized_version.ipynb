```python
import numpy as np
import time
```

```python
#ทดสอบการใช้งานระหว่าง vectorized version และ for loop
#1 inprementing vectorized version
a = np.random.rand(1000000)
b = np.random.rand(1000000)

tic = time.time()
c = np.dot(a,b)
toc = time.time()

print (c)
print ("vectorized version:" + str(1000*(tic-toc))+"ms")
```

```python
#2 การใช้แบบ for loop
c = 0
tic = time.time()
for i in range(1000000):
    c += a[i]*b[i]
toc = time.time()

print (c)
print ("for loop"+ str(1000*(tic-toc))+"ms")
```
